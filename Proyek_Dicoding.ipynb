### Cleaning Data

```python
datetime_columns = ["dteday"]
for column in datetime_columns:
    day_df[column] = pd.to_datetime(day_df[column])
```

```python
day_df.info()
```

```python
datetime_columns = ["dteday"]
for column in datetime_columns:
    hour_df[column] = pd.to_datetime(hour_df[column])
```

## Exploratory Data Analysis (EDA)

### Explore ...

```python
day_df["temp"] = day_df["temp"].apply(lambda x: (x - 32) * 5 / 9)
day_df["temp"]
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: 

```python
temp = day_df["temp"]
register = day_df["registered"]

plt.scatter(temp, register)

plt.title("Perbandingan Suhu dan Jumlah Pendaftaran")
plt.xlabel("Suhu (Derajat Celcius)")
plt.ylabel("Jumlah Pendaftaran")

plt.show()
```

### Pertanyaan 2: 

```python
temp = hour_df["temp"]
register = hour_df["registered"]

plt.scatter(temp, register)

plt.title("Perbandingan Suhu dan Jumlah Pendaftaran")
plt.xlabel("Suhu (Derajat Celcius)")
plt.ylabel("Jumlah Pendaftaran")

plt.show()
```
